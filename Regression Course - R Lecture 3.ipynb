## Question 1: Write your own One-Sample T-Test

```r
my_t_test <- function(x, q=0) {
    alpha = 0.05
    
    dof = length(x) - 1
    mean_x <- mean(x)
    SE <- sd(x) / sqrt(length(x))
    t_value <- mean_x/SE
    t_critical <- unname(quantile(x, prob=1-alpha))
    negative_t_critical <- t_critical * -1
    p_value <- 2*pt(-abs(t_value), dof)
    
    print(paste("Computed t_value: ",t_value))
    print(paste("Degrees of Freedom: ",dof))
    print(paste("p-value: ", format(p_value, scientific = TRUE, digits = 4)))
    print(paste(1-alpha, "percent confidence interval:"))
    print(paste(negative_t_critical,t_critical, sep=","))
    
    if (between(t_value, negative_t_critical, t_critical)){
        print("Failed to Reject Null Hypothesis")
    } else {
        print("Reject Null Hypothesis")
    }
        
}
```

```r
x <- c(2.1, 1.8, 2.3, 2.0, 1.9)
my_t_test(x)
mean(x)
```

```r
t.test(x)
```

## Question 2: Write your own F-Test Functin

```r
my_f_test <- function(x, y) {
    alpha <- 0.05
    
    n_x <- length(x)
    n_y <- length(y)
    dof1 <- n_x - 1
    dof2 <- n_y - 1
    
    var_x <- var(x)
    var_y <- var(y)           
    
    f_stat <- var_x/var_y
    
    
    # Compute p-value (two-tailed)
    if (f_stat > 1) {
        p_value <- 2 * pf(f_stat, dof1, dof2, lower.tail = FALSE)
    } else {
        p_value <- 2 * pf(f_stat, dof1, dof2, lower.tail = TRUE)
    }
    
    # Ensure p-value does not exceed 1
    p_value <- min(p_value, 1)
        
    # Decision rule
    if (p_value < alpha) {
        response <- "Reject Null Hypothesis: Variances are significantly different."
    } else {
        response <- "Fail to Reject Null Hypothesis: Variances are not significantly different."
    }
    
    # Return results as a list
    return(list(
        paste("Computed F-statistic:", round(f_stat, 4)),
        paste("Degrees of Freedom: df1 =", dof1, ", df2 =", dof2),
        paste("p-value:", format(p_value, scientific = TRUE, digits = 4)),
        paste("Response:", response)
    ))
}
```

```r
x <- c(5.2, 4.8, 5.0, 5.1, 4.9)
y <- c(6.1, 5.5, 6.3, 5.9, 6.0, 5.7)

```

```r
my_f_test(x, y)
```

```r
var.test(x, y)
```
